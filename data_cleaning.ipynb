```python
# Custom taking care of special cases
def convert_countries(row):
    if row == "Unspecified" or row == "European Community":
        return np.nan
    elif row == "EIRE":
        return "Ireland"
    elif row == "RSA":
        return "South Africa"
    else:
        return row

# Apply custom function to each element in DataFrame column 'A'
main_df['customer_country'] = main_df['customer_country'].apply(convert_countries)
```
